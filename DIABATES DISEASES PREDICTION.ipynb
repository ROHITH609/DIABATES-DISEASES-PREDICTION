### SPLITING DATASET

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=7)
```

### SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC
```

```python
svm=SVC(C=10,kernel='linear')
```

### TRAINING MODEL

```python
model1=svm.fit(x_train,y_train)
```

```python
y1_pred=svm.predict(x_test)
```

```python
y1_pred
```

### ACCURACY SCORE

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,roc_curve,auc,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y1_pred)
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn=KNeighborsClassifier(n_neighbors=6)
```

```python
model2=knn.fit(x_train,y_train)
```

```python
y2_pred=knn.predict(x_test)
```

```python
accuracy_score(y_test,y2_pred)
```

### XGBClassifier

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
```

```python
model3=xgb.fit(x_train,y_train)
```

```python
y3_pred=xgb.predict(x_test)
```

```python
accuracy_score(y_test,y3_pred)
```

### Logistic Regression

```python
from  sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=1000)
```

```python
model4=lr.fit(x_train,y_train)
```

```python
y4_pred=lr.predict(x_test)
```

```python
accuracy_score(y_test,y4_pred)
```

```python
print(classification_report(y_test,y4_pred))
```

```python
plot_confusion_matrix(model4,x_train,y_train)
```

```python
ptr,ftr,thr=roc_curve(y_test,y4_pred)
```

```python
plt.plot(ptr,ftr)
```
